#### Q3. 
Crie um dataframe `participacao` com as colunas `país`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada país participou, informando se é da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 países de cada competição que mais participou de copas do mundo.

Exemplo de saída (valores inventados):
```
país            copa        num_copas
Brazil          Feminina    8
```

````python
# Criar um dataframe com participações dos times em casa e fora
participacao_time1 = wc[['time_1', 'copa', 'ano']].rename(columns={'time_1': 'país'})
participacao_time2 = wc[['time_2', 'copa', 'ano']].rename(columns={'time_2': 'país'})
participacao_total = pd.concat([participacao_time1, participacao_time2])

# Contar o número de copas distintas por país e tipo de copa
participacao = participacao_total.groupby(['país', 'copa'])['ano'].nunique().reset_index(name='num_copas')

# Exibir o top 5 para Feminina
print("Top 5 países - Copa Feminina:")
display(participacao[participacao['copa'] == 'Feminina'].sort_values(by='num_copas', ascending=False).head(5))

# Exibir o top 5 para Masculina
print("Top 5 países - Copa Masculina:")
display(participacao[participacao['copa'] == 'Masculina'].sort_values(by='num_copas', ascending=False).head(5))
````

#### Q4. 
* Crie um dataframe `gols` com duas colunas `país` e `total_gols` com o total de gols marcados por cada país durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
país        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python
# Calcular gols marcados como time_1
gols_time1 = wc.groupby('time_1')['gols_1'].sum().reset_index().rename(columns={'time_1': 'país', 'gols_1': 'total_gols'})

# Calcular gols marcados como time_2
gols_time2 = wc.groupby('time_2')['gols_2'].sum().reset_index().rename(columns={'time_2': 'país', 'gols_2': 'total_gols'})

# Somar os gols totais por país
gols = pd.concat([gols_time1, gols_time2]).groupby('país')['total_gols'].sum().reset_index()

# Ordenar de forma decrescente
gols = gols.sort_values(by='total_gols', ascending=False)

# Exibir o dataframe
display(gols)
````

#### Q5. 
Qual país tomou mais cartões amarelos somando todas as copas?

Neste exercício você vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observação é uma string represetando uma lista dos cartões amarelos de um único jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16’|1:0|Rosana Gómez', '20’|2:0|Gabriela Chávez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o método genérico `apply` para chamar uma função que você vai criar para transformar uma observação de cartão amarelo em uma contagem de elementos da lista. 

Lembre de levar em consideração que muitas observações são `NaN`. 

Em seguida faça sua mágica para agrupar as informações por país, acumular os cartões de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
país        cartões amarelos
Argentina   97
England     93
Australia   93
...
```

````python
# Função para contar o número de cartões em uma string (considerando NaN)
def contar_cartoes(valor):
    if pd.isna(valor):
        return 0
    # Contar elementos separados por '|' ou considerar como lista
    return len(str(valor).split('|')) if '|' in str(valor) else 1 if valor else 0

# Criar colunas com a contagem de cartões
wc['num_cartoes_1'] = wc['cartao_amarelo_1'].apply(contar_cartoes)
wc['num_cartoes_2'] = wc['cartao_amarelo_2'].apply(contar_cartoes)

# Somar cartões por país (time_1 e time_2)
cartoes_time1 = wc.groupby('time_1')['num_cartoes_1'].sum().reset_index().rename(columns={'time_1': 'país', 'num_cartoes_1': 'cartões amarelos'})
cartoes_time2 = wc.groupby('time_2')['num_cartoes_2'].sum().reset_index().rename(columns={'time_2': 'país', 'num_cartoes_2': 'cartões amarelos'})

# Combinar e somar os cartões totais por país
cartoes = pd.concat([cartoes_time1, cartoes_time2]).groupby('país')['cartões amarelos'].sum().reset_index()

# Ordenar e exibir
cartoes = cartoes.sort_values(by='cartões amarelos', ascending=False)
display(cartoes)
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de pênalti.

Para conseguir essa informação, você precisará trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas também são colunas textuais, onde cada observações apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol está na forma `'Jogador(a) · minuto’'`. Por exemplo:
```
'Alex Morgan · 12’|Rose Lavelle · 20’'
```

Lembre de levar em consideração que muitas observações são `NaN`. 

Recomendo criar um dicionário à parte, onde cada chave será um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente será a contagem de ocorrências desses nomes.

Em seguida basta converter o seu dicionário em um dataframe e imprimí-lo na forma (valores não são inventados 😁):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
# Função para extrair jogadores de uma string de gols
def extrair_jogadores(valor):
    if pd.isna(valor):
        return []
    return [gol.split(' · ')[0] for gol in str(valor).split('|') if ' · ' in gol]

# Criar um dicionário para contar gols por jogador
gols_jogadores = {}

# Processar as colunas de gols
for coluna in ['gols_1_detalhes', 'gols_2_detalhes', 'gols_1_penalti', 'gols_2_penalti']:
    wc[coluna].apply(lambda x: [gols_jogadores.update({jogador: gols_jogadores.get(jogador, 0) + 1}) for jogador in extrair_jogadores(x)])

# Converter o dicionário em um dataframe
df_gols_jogadores = pd.DataFrame(list(gols_jogadores.items()), columns=['jogador(a)', 'num_gols'])

# Ordenar e exibir o top 10
top10_jogadores = df_gols_jogadores.sort_values(by='num_gols', ascending=False).head(10)
display(top10_jogadores)
````
